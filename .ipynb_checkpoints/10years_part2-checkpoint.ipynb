```python
from pymongo import MongoClient
import pandas as pd
from pandas.io.json import json_normalize
import functions as f
import numpy as np
```

```python
client = MongoClient("mongodb://localhost/companies")
```

```python
db = client.get_database()
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
query1 = {"city":{"$in":['New York','San Francisco','Palo Alto','Los Angeles']}}
```

```python
data = list(db["new_comp"].find(query1))
df = pd.DataFrame(data)
```

```python
def fillCoord(df,col):
    col_nulls = df[col].isnull()
    null_df = df[col_nulls]
    idx = null_df.index
    for i in range(len(idx)):
        if df.address1[idx[i]] != None and df.address1[idx[i]] != '':
            df.set_value(idx[i],col,f.getLocation(df.address1[idx[i]]))
    return df
```

```python
fillCoord(df,'location')
```
