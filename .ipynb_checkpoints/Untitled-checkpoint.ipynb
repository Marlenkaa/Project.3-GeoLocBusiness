```python
from pymongo import MongoClient
import pandas as pd
from pandas.io.json import json_normalize
import functions as f
import numpy as np
```

```python
client = MongoClient("mongodb://localhost/companies")
```

```python
db = client.get_database()
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

Creaci√≥n de un DataFrame:

```python
data = list(db["companies"].find())
df = pd.DataFrame(data)
```

Limpieza de la columna 'offices' para obtener las coordenadas:

```python
df = df.explode("offices")
expand = df[["offices"]].apply(lambda r: r.offices ,result_type="expand", axis=1)
df2 = pd.concat([df,expand],axis=1)
df2.head()
```

```python
df2 = df2.drop(columns=["offices"])
```

```python
df2["location"] = df2[["latitude","longitude"]].apply(lambda x:f.coorFormat(x.latitude,x.longitude), axis=1)
```

Filtro para obtener las compa√±√≠as con antiguedad menor a 10 a√±os:

```python
filt = df2['founded_year'] > 2010
new_comp = df2[filt]
new_comp.shape
```

Filtro para quedarnos solamente con las compa√±√≠as en las ubicaciones m√°s frecuentadas:

```python
cities = ['New York','San Francisco','Palo Alto','Los Angeles']

new_comp2 = new_comp.loc[new_comp['city'].isin(cities)]

new_comp2.shape
```
