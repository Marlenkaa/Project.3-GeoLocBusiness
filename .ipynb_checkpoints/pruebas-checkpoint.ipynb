```python
db = client.get_database()
```

```python
a = list(db["companies"].find())
a[:1]
```

```python
df = pd.DataFrame(a)
```

APARTE

Saco solamente columnas de interés para limpiar la localización

```python
companies = list(db.companies.find({},{"name":1,"offices":1,"category_code":1}))
```

Convertimos en pandas para ver mejor los datos

```python
df = pd.DataFrame(companies)
df.head()
```

Separamos los datos contenidos en la columna offices

```python
df = df.explode("offices")
expand = df[["offices"]].apply(lambda r: r.offices ,result_type="expand", axis=1)
df2 = pd.concat([df,expand],axis=1)
df2.head()
```

Elimamos la columna offices

```python
df2 = df2.drop(columns=["offices"])
```
